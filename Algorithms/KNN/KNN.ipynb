### We will be using handwriting dataset for KNN

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape , y_train.shape
X_test.shape , y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
```

```python
#Training the model happens in fit functions(In general)
# no work is happening in training knn model
#All the computation happens while you give the test data
# KNN is BruteForced,So it is one of least choosen while we do classification
model.fit(X_train,y_train)
```

```python
#Testing our model
print(model.predict(X_test[:10]))
print(y_test[:10])
```

```python
model.score(X_test[:100],y_test[:100])
```
