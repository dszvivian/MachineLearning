```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.mixture import GaussianMixture 
from sklearn.cluster import KMeans 

data = pd.read_csv('em9.csv')
f1 = data['V1'].values
f2 = data['V2'].values
X = np.array(list(zip(f1,f2)))
print("x: ",X)
```

```python
print("Graph for whole Dataset")
plt.scatter(f1,f2,c='black')
plt.show()
```

```python
kmeans = KMeans(2)
labels = kmeans.fit(X).predict(X)
print("labels for kmeans:",labels)
centeroids = kmeans.cluster_centers_
print("Centeroids:",centeroids)
```

```python
print("Graph Using KMeans Algorithm")
plt.scatter(f1,f2,c=labels)
```

```python
plt.scatter(centeroids[:,0],centeroids[:,1],marker='*',c='red')
plt.show()
```

```python
gmm = GaussianMixture(2)
labels = gmm.fit(X).predict(X)
print("mean:\n",gmm.means_)
print('\n')
print('Covariances\n',gmm.covariances_)
print("Labels for GMM: ",labels)
```

```python
plt.title("Exception Maximum using GMM")
plt.scatter(f1,f2,c=labels)
plt.show()
```
