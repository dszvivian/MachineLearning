```python
#Read a csv file using dataframe 

import pandas as pd
df = pd.read_csv('customer(A).csv')
print(df)
```

```python
# Plot each column graph vertically 

import matplotlib.pyplot as plt 

income = df['income'].value_counts()
income.plot.bar()
```

```python
# Identify and delete rows that contain duplicate data 

print("Duplicate Rows are :\n",df[df.duplicated(subset=['age','income','student','creditrating','class'])])
print('Resultant DataFrame after deletion:\n',df.drop_duplicates())
```

```python
# Transform categorical columns to integer 

transform = pd.get_dummies(df['income'])
df_new = pd.concat([df,transform],axis=1)
print(df_new)
```

```python
# Transform categorical columns to integer 

df_new['age']=[0 if x=='senior' else 1 for x in df_new['age']]
df_new
```

```python
import numpy as np

new_df = pd.DataFrame({"age_int":np.random.randint(0,100,2)})
labels = ["{0}-{1}".format(i,i+9) for i in range(0,100,10)]
new_df["age_group"]=pd.cut(new_df.age_int,range(0,105,10),right=False,labels=labels)
new_df.head(15)
df_new = pd.concat([df,new_df],axis=1)
print(df_new)
```
