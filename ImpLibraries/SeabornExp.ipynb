### Seaborn:
    .datavisualiztion library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some built in datasets 

```python
#tip dataset
tips = sns.load_dataset('tips') #b default it will convert that to pandas dataframe
```

```python
tips.head()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',style='smoker',size='size')
```

```python
#load iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width')
```
