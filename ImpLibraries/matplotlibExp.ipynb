#### MatplotLib
    - useful to making plots

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
#sine  wave
plt.plot(x,y)
plt.show()

plt.plot(x,z)
plt.show()
```

#### Adding titile x,y axis

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine Variations')
plt.show()
```

#### Parabola

```python
x = np.linspace(-10,10,120)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```
